```python
train_set = df[:2757]
test_set = df[2757:]
```

```python
from prophet import Prophet
```

```python
model = Prophet()
model.fit(train_set)
```

```python
# Create a DataFrame to store the predictions
future = model.make_future_dataframe(periods=len(test_set))
# Make predictions on the future DataFrame
forecast = model.predict(future)

# Extract the relevant columns for evaluation
predicted_prices = forecast['yhat'].tail(len(test_set))
```
